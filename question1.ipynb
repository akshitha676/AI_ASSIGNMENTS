```python
class State:
    def __init__(self, initial,final):
        self.initial = initial
        self.final=final

    def goalTest(self):
        return self.initial==self.final

    def __str__(self):
        return ' '.join(self.initial)

    def __eq__(self, other):
        return isinstance(other, State) and self.initial == other.initial

    def __hash__(self):
        return hash(tuple(self.initial))

    def moveGen(self):
        children = []
        for i in range(len(self.initial)):
            if self.initial[i] == 'R':

                if i + 2 < len(self.initial) and self.initial[i + 2] == '_' and self.initial[i + 1] == 'L':
                    new_config = self.initial[:]
                    new_config[i], new_config[i + 2] = new_config[i + 2], new_config[i]
                    children.append(State(new_config, self.final))
                if i + 1 < len(self.initial) and self.initial[i + 1] == '_':
                    new_config = self.initial[:]
                    new_config[i], new_config[i + 1] = new_config[i + 1], new_config[i]
                    children.append(State(new_config, self.final))

            elif self.initial[i] == 'L':

                if i - 2 >= 0 and self.initial[i - 2] == '_' and self.initial[i - 1] == 'R':
                    new_config = self.initial[:]
                    new_config[i], new_config[i - 2] = new_config[i - 2], new_config[i]
                    children.append(State(new_config, self.final))
                if i - 1 >= 0 and self.initial[i - 1] == '_':
                    new_config = self.initial[:]
                    new_config[i], new_config[i - 1] = new_config[i - 1], new_config[i]
                    children.append(State(new_config, self.final))
        return children


def bfs(start):
    OPEN = [(start, None)]
    CLOSED = []

    while OPEN:
        node_pair = OPEN.pop(0)
        node, parent = node_pair

        if node.goalTest():
            path = reconstructPath(node_pair, CLOSED)
            path.reverse()
            for e in path:
                print(e)
            return

        CLOSED.append(node_pair)
        children = node.moveGen()
        new_nodes = removeSeen(children, OPEN, CLOSED)
        new_pairs = [(c, node) for c in new_nodes]
        OPEN = OPEN + new_pairs

    return []


def dfs(start):
    OPEN = [(start, None)]
    CLOSED = []

    while OPEN:
        node_pair = OPEN.pop()
        node, parent = node_pair

        if node.goalTest():
            path = reconstructPath(node_pair, CLOSED)
            path.reverse()
            for e in path:
                print(e)
            return

        CLOSED.append(node_pair)
        children = node.moveGen()
        new_nodes = removeSeen(children, OPEN, CLOSED)
        new_pairs = [(c, node) for c in new_nodes]
        OPEN = new_pairs + OPEN

    return []


def removeSeen(children, OPEN, CLOSED):
    open_nodes = [node for node, parent in OPEN]
    closed_nodes = [node for node, parent in CLOSED]
    new_nodes = [c for c in children if c not in open_nodes and c not in closed_nodes]
    return new_nodes


def reconstructPath(node_pair, CLOSED):
    path = []
    parent_map = {node: parent for node, parent in CLOSED}

    node, parent = node_pair
    path.append(node)
    while parent is not None:
        path.append(parent)
        parent = parent_map[parent]
    return path


start_config = ['R', 'R', 'R', '_', 'L', 'L', 'L']
goal_config = ['L', 'L', 'L', '_', 'R', 'R', 'R']
start = State(start_config, goal_config)

print("Start State:", start)
print("\n--- BFS Solution ---")
bfs(start)

print("\n--- DFS Solution ---")
dfs(start)
```
